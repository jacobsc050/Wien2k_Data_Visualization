```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Fe.dos1dn', sep="    ", header = 0)
#print(data)
df = pd.DataFrame(data, columns= ["# ENERGY", "total-DOS" ])
fermi_energy = 0.58807
shifted_energy = df["# ENERGY"] - fermi_energy
df["# SHIFTED"] = shifted_energy
print(df)
plt.figure()
df.plot(x = "total-DOS", y = "# SHIFTED", kind = 'line')
plt.show()
"""
with open(r"optic_calc.txt", "r") as file:
    data = file.read()
    print(data)
    file.close()
"""
```

```python
data = pd.read_csv('Fe.epsilonup', sep="  ", skiprows=[0,1,2,3])
print(data)
df = pd.DataFrame(data, columns= ["# Energy [eV]", "Re_eps_xx",  "Im_eps_xx", "Re_eps_zz",  "Im_eps_zz"])
energy_to_wavelength = 1240/df["# Energy [eV]"]
df["wavelength [nm]"] = energy_to_wavelength
df = df[df["wavelength [nm]"]<3000]
plt.figure()
df.plot(x = "wavelength [nm]", y = ["Re_eps_xx", "Im_eps_xx", "Re_eps_zz", "Im_eps_zz"],kind = 'line')
plt.show()
```
