```python
data = pd.read_csv('epsilon.txt', sep="  ", header = 0)
print(data)
df = pd.DataFrame(data, columns= ["# Energy [eV]", "Re_eps_xx", "Im_eps_xx"])

plt.figure()
df.plot(x = "# Energy [eV]", y = ["Re_eps_xx","Im_eps_xx"],kind = 'line')
plt.show()
```

```python
data = pd.read_csv('refraction.txt', sep="  ", header = 0)
print(data)
df = pd.DataFrame(data, columns= ["# Energy [eV]", "ref_ind_xx", "extinct_xx"])

plt.figure()
df.plot(x = "# Energy [eV]", y = ["ref_ind_xx","extinct_xx"],kind = 'line')
plt.show()
```

```python
data = pd.read_csv('dos.txt', sep="    ", header = 0)
fermi_energy = 0.35355
#print(data)
df = pd.DataFrame(data, columns = ["# ENERGY", "1:total"])

shifted_energy = df["# ENERGY"] - fermi_energy
df["# SHIFTED"] = shifted_energy

print(df)

ax = df.plot(x = "# SHIFTED", y = "1:total",kind = 'line')
ax.axvline(0, linestyle = "--",label =  "Fermi Energy", color = "black")
plt.legend(loc = "upper left")
plt.show()
```

```python
data = pd.read_csv('reflectivity.txt', sep="  ", header = 0)

df = pd.DataFrame(data, columns= ["# Energy [eV]", "reflect_xx"])
energy_to_wavelength = 1240/df["# Energy [eV]"]
df["wavelength [nm]"] = energy_to_wavelength
df = df[df["wavelength [nm]"]<3000]
plt.figure()
df.plot(x = "wavelength [nm]", y = ["reflect_xx"],kind = 'line',ylim=[0,1])
plt.rcParams["figure.autolayout"] = True
plt.show()
```
